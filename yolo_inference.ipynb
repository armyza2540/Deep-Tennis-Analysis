```python
# Imports
from ultralytics import YOLO
```

```python
# Making Inference with Trained Model
saved_model_path = './models/yolo5_best.pt'
model = YOLO(saved_model_path)

result = model.predict('./input_videos/input_video.mp4', conf=0.2, save=True)
print(result)
print('Boxes:')
for box in result[0].boxes:
    print(box)
print('Labels:')
for label in result[0].boxes.cls:
    print(label)
```
